```python
#@title Definicion de filtro paso banda para mejorar la senal EEG
import scipy.signal as signal

def filtro_paso_banda(data, lowcut, highcut, fs, order=5):
    b,a = signal.butter(order,[lowcut,highcut],fs=fs,btype='band')
    y = signal.lfilter(b,a,data) # utilizar valores de entrada de la funcion para obtner la senal filtrada
    return y
```

```python
#@title Comparacion respuestas a estimulos objetivos y no objetivos

fig, axs = plt.subplots(4,5,figsize=(10,7))

plot_indexes = [[0,2],[1,2],[2,2],[3,2],[2,1],[2,3],[3,0],[3,4]]

fs = 256
dt = 4
w_len = 125 # Longitud prudente de la ventana para visualizar el P300 (calcular utilizando fs y dt)

for indx,channel in enumerate(channels):
    target_waves = [];        no_target_waves = [];
    no_target_segment = [];   target_segment = [];
    no_target_labels = [];    target_labels = [];

    filtered_signal = filtro_paso_banda(data = time_data[channel],
                                        lowcut = 0.1,
                                        highcut = 10,
                                        fs = fs,
                                        order = 3)

    for n in range(1,len(stims)):
      if (stims[n] - stims[n-1]) == 2: # Valor asignado para estimulos objetivo
        target_waves.append(filtered_signal[n:n+w_len])
      if (stims[n] - stims[n-1]) == 1: # Valor asignado para estimulos no objetivo
        no_target_waves.append(filtered_signal[n:n+w_len])

    targets_ = np.array(target_waves)
    no_targets_ = np.array(no_target_waves)

    ax = axs[plot_indexes[indx][0],plot_indexes[indx][1]]

    tiempo = np.arange(0, no_targets_.shape[1]*dt, dt) # Actualmente en segundos, convertir a milisegundos para encontrar el valor "P300"
    polarizacion = -1 # Definir si debe ser 1 o -1

    ax.plot(tiempo,no_targets_.mean(axis=0)*polarizacion,label='Noise_mean')
    ax.plot(tiempo,targets_.mean(axis=0)*polarizacion,label='Target_mean')

    ax.vlines(x = None, ymin = -2, ymax = 2, colors = 'red', label = 'pre-P300')
    ax.vlines(x = None, ymin = -2, ymax = 2, colors = 'red', label = 'Post-P300')

    ax.set_title('Channel '+channel)

for m in range(4):
    for n in range(5):
        if [m,n] not in plot_indexes:
            axs[m,n].cla()
            axs[m,n].axis('off')

plt.tight_layout()
```
